```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import Utils.greeks as gr
import Utils.option_pricing_utils as opu
```

```python
S0 = 20000
dS = np.arange(-0.05*S0, 0.05*S0, 0.01)
dt = 1/365
r = 0.02
sigma = 0.8
q = 0
T = 0.2
K = 20000
type = 'c'
```

```python
gamma = gr.get_gamma(S0, K, T, sigma, r, q, type)
theta = gr.get_theta(S0, K, T, sigma, r, q, type)
```

```python
delta_hedged_portfolio_value = 1/2*gamma*dS**2 + theta*dt
breakeven_stock_move = np.sqrt(-2*theta*dt/gamma)
breakeven_stock_move
```

```python
fig, ax = plt.subplots()
ax.plot(dS, delta_hedged_portfolio_value)
plt.axhline(y=0.0, color='r', linestyle='-')
plt.axvline(x=breakeven_stock_move, color='g', linestyle='-')
plt.axvline(x=-breakeven_stock_move, color='g', linestyle='-')
plt.show()
```
